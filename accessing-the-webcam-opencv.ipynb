**Using this code you can open your default webcam.**

**Use this code on your python IDE software**

```python
import cv2
import sys

# Get camera/video source (default 0 for primary camera)
s = 0
if len(sys.argv) > 1:
  s = sys.argv[1]

source = cv2.VideoCapture(s)

# Window name and creation
win_name = 'Camera Preview'
cv2.namedWindow(win_name, cv2.WINDOW_NORMAL)  # Now should work!

# Video capture loop
while cv2.waitKey(1) != 27:  # Escape
  has_frame, frame = source.read()
  if not has_frame:
    break
  cv2.imshow(win_name, frame)

# Cleanup
source.release()
cv2.destroyWindow(win_name)
```
