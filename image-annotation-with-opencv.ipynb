# Annotating Images
In OpenCV, annotations typically refer to adding text, lines, shapes, or other graphical elements to images for visualization or analysis purposes. Here's how we can perform annotation in OpenCV:

1. Draw lines
1. Draw circles
1. Draw rectangles
1. Add text

When we wish to demonstrate our programme or mark your findings for presentations, these come in handy. When developing and debugging, annotations might be helpful as well.

```python
#importing Libaries

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# Reading an image

rice_image = cv.imread("../input/rice-field/rice field.jpg",cv.IMREAD_COLOR)

# Display the original image
plt.imshow(rice_image[:, :, ::-1])
```

# Drawing Line 
> Syntax : img = cv2.line(img, pt1, pt2, color[, thickness[, lineType[, shift]]])

The function has 4 required arguments:

* img: Image on which we will draw a line
* pt1: First point(x,y location) of the line segment
* pt2: Second point of the line segment
* color: Color of the line which will be drawn

Other optional arguments that are important for us to know include:

* thickness: Integer specifying the line thickness. Default value is 1.

* lineType: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.


```python
imageLine = rice_image.copy()

# The line starts from (200,200) and ends at (500,100)
# The color of the line is YELLOW (Recall that OpenCV uses BGR format)
# Thickness of line is 6px
# Linetype is cv.LINE_AA

cv.line(imageLine, (200, 200), (500, 100), (0, 255, 255), thickness=6, lineType=cv.LINE_AA);

# Display the image
plt.imshow(imageLine[:,:,::-1])
```

# Circle Drawing

> **Syntax :** img = cv2.circle(img1, center, radius, color[, thickness[, lineType[, shift]]])

* img1: Image on which we will draw a line

* center: Center of the circle

* radius: Radius of the circle

* color: Color of the circle which will be drawn

(optional) arguments which we are going to use quite extensively.

* thickness: Thickness of the circle outline (if positive). If a negative value is supplied for this argument, it will result in a filled circle.

* lineType: Type of the circle boundary. This is exact same as lineType argument in cv2.line

```python
spray = cv.imread("../input/rice-field/spray.jpg")
spray = cv.cvtColor(spray,cv.COLOR_BGR2RGB)

plt.imshow(spray)
```

```python
imageCircle = spray.copy()

cv.circle(imageCircle, (420,150), 100, (0, 0, 255), thickness=5, lineType=cv.LINE_AA);

plt.imshow(imageCircle[:,:,::-1])
```

# Rectangle Drawing

> **Syntax:** img = cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])

* img: Image on which the rectangle is to be drawn.
* pt1: Vertex of the rectangle. Usually we use the top-left vertex here.
* pt2: Vertex of the rectangle opposite to pt1. Usually we use the bottom-right vertex here.
* color: Rectangle color

Optional

* thickness: Thickness of the circle outline (if positive). If a negative value is supplied for this argument, it will result in a filled rectangle.

* lineType: Type of the circle boundary. This is exact same as lineType argument in cv2.line

```python
imageRectangle = spray.copy()

cv.rectangle(imageRectangle, (370, 100), (440, 220), (150, 0, 150), thickness=5, lineType=cv.LINE_8)

# Display the image
plt.imshow(imageRectangle[:, :, ::-1])
```

# Text In Image

> **Syntax:** img = cv2.putText(img, text, org, fontFace,fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])

* img: Image on which the text has to be written.
* text: Text string to be written.
* org: Bottom-left corner of the text string in the image.
* fontFace: Font type
* fontScale: Font scale factor that is multiplied by the font-specific base size.
* color: Font color

Optional

* thickness: Integer specifying the line thickness for the text. Default value is 1.
* lineType: Type of the line. Default value is 8 which stands for an 8-connected line. Usually, cv2.LINE_AA (antialiased or smooth line) is used for the lineType.

```python
imagetext = spray.copy()
text = "A Farmaer is Spraying Pesticide In His Crop Field"
fontScale = 2
fontFace = cv.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 255)
fontThickness = 2

cv.putText(imagetext, text, (45, 50), fontFace, fontScale, fontColor, fontThickness, cv.LINE_AA);

# Display the image
plt.imshow(imagetext[:, :, ::-1])
```
